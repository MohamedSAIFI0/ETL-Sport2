```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\P15\Desktop\ETL\data\filtered_data.csv")
print(df.head())
```

```python
plt.plot(df["Sport"],df["Games Played"])
plt.xlabel("Sport")
plt.ylabel("Games Played")
plt.title("Games Played in each sports")
plt.show()
```

```python
plt.bar(df["Sport"],df["Goals"],color="red")
plt.xlabel("Sport")
plt.ylabel("Goals")
plt.title("Goals in every sport")
plt.show()
```

```python
plt.scatter(df["Sport"],df["Assists"],color="green")
plt.xlabel("Sport")
plt.ylabel("Assists")
plt.title("Assists in every sport")
```
